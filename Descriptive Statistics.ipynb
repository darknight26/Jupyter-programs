```python
def skewness(x):
    res = 0
    mu = x.mean()
    s = x.std()
    for i in x:
        res+=(i-mu)*(i-mu)*(i-mu)
    res/=len(x)*s*s*s
    return res
print("Skewness of male population ",skewness(ml2_age))
print("Skewness of male population ",skewness(fm2_age))
```

```python
def pearson(x):
    return 3*(x.mean()-x.median())*x.std()
print("Pearson skewness of male ",pearson(ml2_age))
print("Pearson skewness of female ",pearson(fm2_age))
```
