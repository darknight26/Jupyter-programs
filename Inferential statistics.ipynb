```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
data = pd.read_csv(r"C:\Users\ashut\Downloads\2013_ACCIDENTS_CAUSES_GU_BCN_2013.csv", encoding= 'unicode_escape')
```

```python
data
```

```python
data['Date']= data[u'Dia de mes'].apply(lambda x: str(x))+'-' + data[u'Mes de any'].apply(lambda x: str(x))+'-'+'2013'
```

```python
data['Date']=pd.to_datetime(data['Date'])
```

```python
data
```

```python
accidents = data.groupby(data['Date']).size()
print(accidents)
accidents.mean()
```

```python
import math
import matplotlib.pylab as plt
fig, ax = plt.subplots(1,1 , figsize=(12, 4))
plt.ylabel("Number of accidents")
plt.xlabel("Day")
plt.plot(range(0,365),np.array(accidents),'b-+',lw=0.7,alpha=0.7)
plt.plot(range(0,365),[accidents.mean()]*365,'r',lw=0.9,alpha=0.9)
plt.show()
```

```python
fig, ax = plt.subplots(1,1 , figsize=(12, 4))
plt.ylabel("Frequency")
plt.xlabel("Number of Accidents")
plt.hist(np.array(accidents),bins=20)
ax.axvline(x=accidents.mean(),color=[1,0,0])
plt.show()
```

```python
df = accidents.to_frame()
m=[]
for i in range(10):
    df['for_testing']=False
    sampled_ids=np.random.choice(df.index,size=np.int64(np.ceil(df.index.size*0.25)),replace=False)
    df.loc[sampled_ids,"for_testing"]=True
    accidents_sample=df[df["for_testing"]==True]
    m.append(accidents_sample[0].mean())
    print("Sample" + str(i) +": Mean = ","%0.4f"%accidents_sample[0].mean())
```

```python
fig,ax = plt.subplots(1,1,figsize=(12,2))
x=range(10)
ax.step(x,m)
plt.xlabel("Sample Number")
plt.ylabel("Mean")
```

```python
fig,ax = plt.subplots(1,1,figsize=(12,2))
x=range(10)
ax.step(x,m,where='mid')
ax.set_xlabel("Sample Number")
ax.set_ylabel("Mean")
```

```python
df = accidents.to_frame()
m=[0]*10000
for i in range(10000):
    df['for_testing']=False
    sampled_ids=np.random.choice(df.index,size=200,replace=False)
    df.loc[sampled_ids,"for_testing"]=True
    accidents_sample=df[df["for_testing"]==True]
    m[i]=accidents_sample[0].mean()

fig,ax= plt.subplots(1,1,figsize=(12,3))
plt.hist(np.array(m),bins=50)
plt.ylabel("frequency")
plt.xlabel("Mean Value of 10000 Samples of size 200")
ax.axvline(np.array(m).mean(),ymin=0,ymax=700,color=[1,1,0])
plt.savefig("empiricalmean.png",dpi=300, bbox_inches='tight')
plt.show()
print("Sample_mean : ",np.array(m).mean())
```

```python
#to check Laplace Standard Error 
rows= np.random.choice(df.index,size=200)
sample_df=df.loc[rows]
sig_mean=sample_df.std()/math.sqrt(200)
print('Direct estimation of SE from one sample of 200 elements:', \
       sig_mean[0])
print('Estimation of the SE by simulating 10000 samples of 200 elements:',  \
       np.array(m).std())
```
